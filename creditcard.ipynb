```python
df['Category'] = df['Category'].fillna('unknown')
df['Payment_Mode'] = df['Payment_Mode'].fillna('unknown')
df['Transaction_Status'] = df['Transaction_Status'].fillna('unknown')
df['Location'] = df['Location'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
# convert str from  float columns
df['Amount'] = df['Amount'].astype(float)
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median)
df.isnull().sum()
```

```python
#convert transaction date into datetime format and extract year ,month,day as new columns
print(df.head()) #print 5 rows
print(df.dtypes) #check datatype

df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
print(df.dtypes) #check datatype
```

```python
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Date'] = df['Transaction_Date'].dt.day

df.head()
#check info
df.info()
```

```python
#retrieve all transactions made in january 2023
jan_2023 = df[(df['Year'] == 2023) & (df['Month'] == 2)]
jan_2023.head()

#create a report
jan_2023.to_csv("feb_2023.csv")
```

```python
#Find transactions where Amount > 1000 and Transaction_Type is "Online".
high_online_t = df[(df['Amount'] > 1000) & (df['Transaction_Type'] == 'Online')]
print(high_online_t.head())
```

```python
#Select only Approved transactions from the dataset.
approved_t = df[(df['Transaction_Status'] == 'Approved')]
print(approved_t.head())
print(approved_t.shape)
```

```python
# feature engineering = trick for column selecting
#apply
#lambda
#cut
#bins
```

```python
#data Manipulation and Feature Engineering
#Create a new column Discounted_Amount, assuming a 5% discount on all transactions above $500
df['Discounted_Amount'] = df['Amount'].apply(lambda x : x * 0.95 if x > 4000 else x)
df[['Amount','Discounted_Amount']].head()
```

```python

#Categorize the Transaction_Amount into Low, Medium, and High based on:
#Low: Below $2000
#Medium: Between $2000 - $4000
#High: Above $4000

#Use cut when you need to segment and sort data values into bins. 
# This function is also useful for going from a continuous variable to a categorical variable. 
# For example, cut could convert ages to groups of age ranges. Supports binning into an equal number of bins,
#  or a pre-specified array of bins.

df['Amount_category'] = pd.cut(
    df['Amount'],
    bins=[-float('inf'),2000,4000,float('inf')],
    labels=['low','medium','high']
)
df[['Amount','Amount_category']].tail()
```

```python
#Drop the Merchant column if more than 30% of values are missing.
df.isnull().sum()
missing_value = df['Merchant'].isnull().mean()
if missing_value > 0.3:
    df.drop(columns=['Merchant'])
```

```python
#Aggregation and Insights
#Find the total transaction amount per Category
total_amt_cat = df.groupby('Category')['Amount'].sum().reset_index()
total_amt_cat
```

```python
#Q1 find the avg transaction amount per category

avg_trans_cat = df.groupby('Category')['Amount'].mean().reset_index()
print(avg_trans_cat)
```

```python
#Q2 find the count transaction amount per category

count_amt_cat = df.groupby('Category')['Amount'].count().reset_index()
count_amt_cat
```

```python
#Q3 Determine the number of declined transactions per Payment_Mode.

declined_trans_payment_mode = df[df['Transaction_Status'] == 'Declined'].count()
print(declined_trans_payment_mode['Payment_Mode'])
```

```python
#Q4 Identify the top 5 most frequent merchants based on transaction count.
merchant_count = df['Merchant'].value_counts().head(5).reset_index()
merchant_count.columns = ['Merchant','Amount']
print(merchant_count)
```

```python
#Q5 Find the average transaction amount per Location
avg_trans_loc = df.groupby('Location')['Amount'].mean().reset_index()
avg_trans_loc.head()
```

```python
#Find customers who made more than 10 transactions in a single day (potential fraud).
df['Date'] = df['Transaction_Date'].dt.date   #made new column
transactions_per_day = df.groupby(['Customer_ID','Date']).size().reset_index(name='Transaction_Count')
potential_fraud_customers = transactions_per_day[transactions_per_day['Transaction_Count'] > 1]
potential_fraud_customers
```
