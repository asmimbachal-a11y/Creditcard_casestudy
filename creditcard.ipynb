```python
import pandas as pd
```

```python
#read CSV
df = pd.read_csv("https://raw.githubusercontent.com/Pankaj-Str/Complete-Python-Mastery/refs/heads/main/53%20DataSet/credit_card_transactions.csv")
df.head()
 
# other method = copy the link from raw data and paste it at the place of the name of that file in the above code  
```

```python
#check size ie. no of rows and columns
print(df.shape)
```

```python
#print only columns name
df.columns.to_list()
```

```python
#to check the summary STATISTICS
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Category'] = df['Category'].fillna('unknown')
df['Payment_Mode'] = df['Payment_Mode'].fillna('unknown')
df['Transaction_Status'] = df['Transaction_Status'].fillna('unknown')
df['Location'] = df['Location'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
# convert str from  float columns
df['Amount'] = df['Amount'].astype(float)
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median)
df.isnull().sum()
```

```python
#convert transaction date into datetime format and extract year ,month,day as new columns
print(df.head()) #print 5 rows
print(df.dtypes) #check datatype

df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
print(df.dtypes) #check datatype
```

```python
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Date'] = df['Transaction_Date'].dt.day

df.head()
#check info
df.info()
```

```python
#retrieve all transactions made in january 2023
jan_2023 = df[(df['Year'] == 2023) & (df['Month'] == 2)]
jan_2023.head()

#create a report
jan_2023.to_csv("feb_2023.csv")
```

```python
#Find transactions where Amount > 1000 and Transaction_Type is "Online".
high_online_t = df[(df['Amount'] > 1000) & (df['Transaction_Type'] == 'Online')]
print(high_online_t.head())
```

```python
#Select only Approved transactions from the dataset.
approved_t = df[(df['Transaction_Status'] == 'Approved')]
print(approved_t.head())
```
